````python
import ast
import pandas as pd
df = pd.DataFrame.from_dict(ast.literal_eval(response.text))
df
````

> Como o texto não tem nenhuma palavra reservada do Python, o ``ast.literal_eval`` funcionou muito bem, mas, caso no meio de seu texto tenha ``True``, ``False``, ``None`` ou qualquer outra palavra reservada, você terá que tratar antes.

# <font color=green> 3. CONCEITUAÇÃO BÁSICA SOBRE INTERNET E HTML

> A Terceira Lei de Newton (Princípio da Ação e Reação) diz que para toda força de ação existe uma força de reação.
>
> Sei que parece filosófico mas é o mesmo mecanismo que rege (a grosso modo) a internet.

> Imagine a seguinte ação...

<center><img width="50%" src="img/google.png"></center>

> O que acontece por tras dos panos é isso...

<center><img width="100%" src="img/request.png"></center>

> O servidor retorna um script, contendo instruções de como montar a página e mostrar para você.
> 
> Ai seu dispositivo pega esse ``pacote de peças`` e cria o ``quebra cabeças`` resultado na sua tela.

<center><img width="60%" src="img/quebra.png"></center>

> A resposta recebida pelo servidor é um script html, contendo as instruções de montagem da página.
>
> HTML abreviação para a expressão inglesa HyperText Markup Language, que significa: "Linguagem de Marcação de Hipertexto"
>
> <font color='#DC143C'> Ou seja, não é uma *linguagem de programação*
>
> <font color='black'> Ela usa **tags** para poder construir aquilo que vimos quando acessamos uma página

> Vamos ver esse exemplo: **/arquivos_apoio/HtmlBasico.html**

> Tags "arroz de festa"
>
> - ```<div>``` - define as divisão da página.
>
> - ```<h1>, <h2>, <h3>, <h4>, <h5>, <h6>``` - marcadores de títulos.
>
> - ```<p>``` - marcador de parágrafo.
>
> - ```<a>``` - hiperlink.
>
> - ```<img>``` - exibição de imagens.
>
> - ```<table>``` - definição de tabelas.
>
> - ```<ul>, <li>``` - definição de listas.

````python
url = 'https://pt.wikipedia.org/wiki/Ayrton_Senna'
````

````python
# Como você já importou o Iframe, não precisa fazer isso novamente
IFrame(url, width='100%', height=400)
````

````python
# Abrindo a página em memória
response = urlopen(url)

# vai retornar um simples objeto
response
````

````python
# Vamos ler esse objeto 
html = response.read()
html
````

````python
# Então vamos tentar ler e...
soup = bs4.BeautifulSoup(html, 'html.parser')
soup
````

````python
#Está feio? Não tem problema!
print(soup.prettify())
````

> A página tem um título, e queremos pegá-lo.
>
> Para isso basta identificarmos a TAG HTML através do "inspetor de elementos (F12)" e podemos ver que...

<center><img width="100%" src="img/ayrton.png"></center>

> Nesse caso, basta informar o TAG (marcação) HTML que se deseja, e pronto!

````python
soup.find('h1').get_text()
````

> Mas é aqui que mora o **problema**.
>
> Esse caso foi fácil, mas muitos programadores *enfiam* H1 até na barra da calça, ai temos que usar de conhecimento, o qual, vou trazer nesse treinamento
>
> Vamos ver a seguir quantos H1 temos nessa página

````python
# Procure todas TAG H1 na página
soup.findAll('h1')
````

> Vou querer mostrar na tela o conteúdo de todas tags ``h3``, e vou querer mostrar com um divisor entre elas, para isso, vou te mostrar o laço de repetição ``for`` em um exemplo simples:

````python
h3_exemplo = ['eu','sou','feliz']

# Quantos elementos tem a lista
len(h3_exemplo)
````

````python
for elemento in h3_exemplo:
    print('má oi')
````

````python
for elemento in h3_exemplo:
    print(elemento)
````

````python
for elemento in h3_exemplo:
    print('-'*100,'\n',elemento)
````

> Te mostrei sem dor, um pouco de lista e o laço ``for`` em ação.
> Legal é mostrar como criar lista de compras, lista de frutas...mas usar a lista no contexto da vida real é melhor ainda

> Agora vamos aplicar esse conceito na extração de informação da página

````python
#Se fosse H2 teria muito mais elementos
for i in soup.findAll('h3'):
    print('-'*100,'\n',i)
````

> Mas os criadores do BS4 nos amam, e criaram o ``get_text`` e veja que delicinha (como diz 'filipe deschamps')

````python
#Se fosse H2 teria muito mais elementos
for i in soup.findAll('h3'):
    print(i.get_text())
````
