### 1.2 Evitando erros de links e respostas

```python
# Importando bibliotecas
from urllib.error import URLError, HTTPError
```

> Vamos fazer agora a consulta de nossa página na WEB, mas vamos nos atentar a uns pontos extra (possiveis erros)

```python
# endereço
url = "https://www.camaraindaial.sc.gov.br/pg/proposicoes"

# Carregue o agente que será usado na requisição
headers = {'User-Agent': agente}

# Efetuando a consulta SE NÃO TIVER ERRO NO LINK
try:
    req = Request(url, headers = headers)
    response = urlopen(req)
    print(response.read())
    
except HTTPError as e:
    print(e.status, e.reason)
    
except URLError as e: 
    print(e.reason)
```

> No próximo teste, colocamos uma url de um dominio existente mas página não existe

```python
# Endereço
url = "https://www.camaraindaial.sc.gov.br/joviano" 

# Carregue o agente que será usado na requisição
headers = {'User-Agent': agente}

# Efetuando a consulta SE NÃO TIVER ERRO NO LINK
try:
    req = Request(url, headers = headers)
    response = urlopen(req)
    print(response.read())
    
except HTTPError as e:
    print('excessao 1 -> url não localizada')    
    print(e.status, e.reason)
    
except URLError as e: r
    print('excessao 2')
    print(e.reason)
```

> Nesse teste estamos consultando um dominio e uma página que não existem

```python
# Endereço
url = "https://www.camaraindaial.ZZ.gov.br/joviano" 

# Carregue o agente que será usado na requisição
headers = {'User-Agent': agente}

# Efetuando a consulta SE NÃO TIVER ERRO NO LINK
try:
    req = Request(url, headers = headers)
    response = urlopen(req)
    print(response.read())
    
except HTTPError as e:
    print('excessao 1 -> dominio existe, pagina não')    
    print(e.status, e.reason)
    
except URLError as e: 
    print('excessao 2 -> dominio não existe')
    print(e.reason)
```

> Fizemos três vezes o mesmo código, e aqui quero ensinar vocês a criar uma função, reduzindo assim o código nas futuras consultas

```python
def ConsultaWeb(url):
    try:
        req = Request(url, headers = headers)
        response = urlopen(req)
        print(response.read())

    except HTTPError as e:
        print('excessao 1 -> dominio existe, pagina não')    
        print(e.status, e.reason)

    except URLError as e: 
        print('excessao 2 -> dominio não existe')
        print(e.reason)    
```

```python
# Olha como fica mais fácil
ConsultaWeb('https://www.camaraindaial.sc.gov.br/pg/proposicoes')
```

> Podemos resumir os erros para um ``pass``, aonde quando não possivel acessar a url, ao invés de explodir um erro, ele ``passa``

```python
def ConsultaWebB(url):
    try:
        req = Request(url, headers = headers)
        response = urlopen(req)
        return response.read()

    except:
        pass
```

```python
# Tenta com um erro, se não der, passa...
ConsultaWebB('https://www.camaraindaial.sc.gov.br/joviano')
```

```python
# Tenta com outro erro, se não der, passa...
ConsultaWebB('https://www.camaraindaial.ZZ.gov.br/joviano')
```

```python
# Se o link estiver OK, ai sim os dados são retornados
ConsultaWebB('https://www.camaraindaial.sc.gov.br/pg/proposicoes')
```

### 1.3 Limite tempo resposta

> Anteriormente eu falei que **não existe um único caminho** tampouco aquele que **seja o único correto**. Lembra?
>
> Pois bem, vamos usar um jeitinho diferente para mostrar a mesma captura feita até então, mas agora com timeout.

> Podemos usar um parâmetro do ``requests`` que limita o tempo de resposta, ou seja, se o site não responder em ``x`` tempo, retorna erro. 

```python
import requests
```

```python
def ConsultaWebTO(url):
    try:
        req = requests.get(url, headers = headers, timeout = 1) 
        return req.content      

    except:
        pass
```

```python
ConsultaWebTO('https://www.camaraindaial.sc.gov.br/pg/proposicoes')
```

> Vamos resumir então e criar uma função para reutilizar isso depois?

```python
def Cabecalho(html):
    dt = html.find_all('dt')
    dd = html.find_all('dd')
    dic = {}
    for i in range(len(dt)):
        x = dt[i].get_text()
        y = dd[i].get_text()
        dic[x] = y
    return dic
```

```python
def TabelaProposicoes(quantidade, ano):
    lista = []
    for i in range(quantidade):
        lista = lista + [ProposicaoCabecalho(i+1,ano)]
    
        print(i+1,'consultas efetuadas | aguardando 0.5 segundos')
        time.sleep(0.5) # meio segundo de espaço entre cada consulta [você vai ter que dizer qual o melhor]

    tabela = pd.DataFrame(lista)
    return tabela
```

```python
def TabelaProposicoesRange(inicia, quantidade, ano):
    lista = []
    for i in range(quantidade):
        lista = lista + [ProposicaoCabecalho(i+1   +inicia-1   ,ano)]
    
        print(i+1,'consultas efetuadas | aguardando 0.5 segundos')
        time.sleep(0.5) # meio segundo de espaço entre cada consulta [você vai ter que dizer qual o melhor]

    tabela = pd.DataFrame(lista)
    return tabela
```

### 3.2 Laço ``while`` (enquanto)

> Esse laço é mais "nojento", pois pode fazer seu algoritmo entrar em **LOOP INFINITO**

> Imagine a regra textual:
>
> Enquando x > 10 OU x < 55 então faça tal coisa...

```python
x = -99999999
x > 10 or x < 55
```

```python
x = 99999999
x > 10 or x < 55
```

```python
x = 0
x > 10 or x < 55
```

```python
x = 20
x > 10 or x < 55
```

<center><img width="100%" src="img/while.png"></center>

```python
# parametros gerais da consulta
inicar_em = 15
quantidade = 10
ano = 2020
ultima_consulta = inicar_em + quantidade - 1

# variaveis para loop
lista = []
i = 1

while inicar_em <= ultima_consulta:

    lista = lista + [ProposicaoCabecalho(inicar_em,ano)]

    print(i,'consultas efetuadas | aguardando 0.5 segundos')
    time.sleep(0.5)

    # carregando incremental as variáveis, com dois "jeitos" diferentes
    inicar_em = inicar_em+1 
    i+=1
          
        
tabela = pd.DataFrame(lista)
tabela
```

### 3.2.1 "Escureceu!" E o que devo fazer agora?

> Vamos acessar o [site da câmara](https://www.legislador.com.br//LegisladorWEB.ASP?WCI=ProposicaoConsulta&ID=3&tpProposicao=1&) e verificar qual é a ultima proposição de hoje.
>
> Então...

```python
# parametros gerais da consulta
inicar_em = 805
quantidade = 10
ano = 2021
ultima_consulta = inicar_em + quantidade - 1

# erros
erros = 0
erros_admissiveis = 2

# variaveis para loop
i = 1
lista = []

while inicar_em <= ultima_consulta and erros <= erros_admissiveis:

    try:
        x = ProposicaoCabecalho(inicar_em,ano)
        lista = lista + [ProposicaoCabecalho(inicar_em,ano)]
    except:
        erros += 1
        pass

    print(i,'consultas efetuadas | aguardando 0.5 segundos')
    time.sleep(0.5)

    # carregando incremental as variáveis, com dois "jeitos" diferentes
    inicar_em = inicar_em+1 
    i+=1
          
        
tabela = pd.DataFrame(lista)
tabela
```
