```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def PegaHtmlLocal(nome, path_subpasta = ''):

    # trata path do sublocal
    if path_subpasta == '':
        path_subpasta = ''
    else: 
        path_subpasta = path_subpasta + '/'
    
    # abra o arquivo, e alimente variável.
    # o 'r' significa que é somente para ler o arquivo.
    with open(path_subpasta + str(nome) + '.html', 'r') as f:
        soup = BeautifulSoup(f, 'html5lib')

    return soup
```

```python
PegaHtmlLocal(1,'bkp_site')
```

```python
lista = []

for i in range(5):
    lista = lista + [PegaHtmlLocal(i+1,'bkp_site')]
    
lista
```
